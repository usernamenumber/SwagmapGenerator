## Proof of Concept: A Swagmap Solver / Generator
Suppose you have a library of training content that...

* Is divided into short, single-subject projects, all indexed by:
  * Provided skills: Skills this lesson teaches
  * Required skills: Skills this lesson assumes knowledge of (and can be used to assess)
  
Granted that's no easy feat to set up, but part of the purpose of this document is to show how it can be worthwhile.

Here's a simple set of basic programming exercises we can use as an example. The list is formatted in [YAML](https://en.wikipedia.org/wiki/YAML), which has the advantage of being easily readably by humans and software alike.

Note the "questionably designed project" in particular. That's going to be relevant later on.


```python
swagifacts_yml = """
---
Hello World:
  Provides: 
    - Print a string
    
Hello X (custom msg):
  Requires: 
    - Print a string
  Provides:
    - Variables
    
What's Your Name?:
  Requires:
    - Print a string
  Provides:
    - Prompt for input
    
Say Hello 10 Times:
  Requires:
    - Print a string
  Provides: 
    - While
    
Solve Simple Math Problems:
  Provides:
    - Arithmetic operators
    
Roll the Dice:
  Requires:
    - Variables
    - Arithmetic operators
  Provides:
    - If / else
    - random.rand
    
A Questionably Designed Project:
  Requires:
    - Network sockets
    - 3D modeling
    - Music composition
  Provides:
    - Arithmetic operators
    
Simple Menu:
  Requires:
    - Print a string
    - Prompt for input
    - While
    - If / else
"""
```

### Just the (Swagi)facts
This list can be turned into a library of Swag artifacts, aka "***swagifacts***". A swagifact is an accomplishment for which students can be given credit in a [learning record store](https://en.wikipedia.org/wiki/Learning_Record_Store) (in this case, completion of each of the lessons described above constitutes a swagifact).

### The swagmap, visualized: Swagifacts assemble! 
The swagifacts can now be assembled into a structured collection called a ***swagmap***, which can be rendered into a nice visualization courtesy of the [igraph](http://igraph.org/python/) graphing library.

The blue nodes in the image below represent projects, and the small green nodes represent the skills each project provides, and  on which other projects depend. 

The *START* node represents zero knowledge of the subject

```python
import swagmap_generator as sg ; reload(sg)
projects = sg.ProjectLibrary.from_yaml(swagifacts_yml)
projects.render_graph()
```

### Where do we go from here?
The fancy image is pretty cool, and if it was all we had it could still be a helpful tool for assembling curricula, but the real beauty here is that once you have everything in a graph, you can use pathfinding algorithms to calculate the best path between two points automatically.

What if we wanted to assemble a learning plan from our library (and remember that this library could contain hundreds of swagifacts covering a huge breadth of subjects, only a fraction of which are relevant to our intro programming students) that represents the optimal path for someone who wants to go from zero subject matter knowledge to being able to do the "*Simple Menu*" exercise, spending as little time as possible on topics that aren't relevant to the task?

For this example we'll use a method called `build_path_to`, which, given the name of an exercise, finds the optimal path through the swagmap to learn it. 

### One step at a time...
The first time through we're going to pass `max_loops=2` as an argument, which will make the process stop partway through so we can learn more about how it works.

Here's a basic overview of what's about to happen:

1. Find the skills required to complete the *Simple Menu* project. 
1. For each of those skills, find the project with the most direct path from zero knowledge that teaches that skill.
1. Do these new projects have their own requirements? Are those requirements met by the current set of projects? If not, repeat the process for those projects.

For now, because of the `max_loops` setting it will only go through the steps once.

Projects that have been incorporated into the learning plan are colored green. One of these is the *Roll the Dice* project, which expects the student to already know about arithmetic operators ("+", "-", etc). Because this topic isn't covered by any of the selected projects (yet), it is colored red and will be resolved in the next pass. 

```python
selected = projects.build_path_to("simple menu",max_loops=2)
projects.render_graph()
```

### Finding Resolution
Note that there are two projects that teach about arithmetic in programs: *Solve Simple Math Problems*, and that "questionably designed" project. 

*Solve Simple Math Problems* is definitely the better choice, because all of its dependencies are already met, whereas the other project brings in a bunch of things that aren't relevant to the *Simple Menu* project, but will the pathfinder know this?

With each iteration of the search, paths that are attached to the selected projects are "weighted" so they will be preferred by the pathfinding algorithm... so yes, it will!

I haven't actually tested how necessary the weighting even is yet, or what the best approach is, but as you're about to see, it works. :) Note that the last arrow from *arithmetic operators* to *Roll the Dice* may still show up as gray instead of green; this is a cosmetic bug I haven't yet fixed, but the dependencies for *Roll the Dice* have still been met.

```python
selected = projects.build_path_to("simple menu")
projects.render_graph()
```
